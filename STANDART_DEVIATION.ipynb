### STANDART DEVIATION

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import OrderedDict
import math

%pylab inline
```

```python
data = pd.read_csv('GEC_Jan_2009.dat', '\s+', header = None, parse_dates = True)
data.columns = ['Time', 'GEC' ]
plot(data.Time, data.GEC)
N = len(data.GEC)
```

```python
d_avGEC = np.zeros(2*31).reshape(2, 31)
k = 0
for i in range(31):
    d_avGEC[0,i] = i + 1
    d_avGEC[1,i] = mean(data.GEC[k:k+12])
    k = k+12
print(d_avGEC)
```

```python
fig1,ax1=plt.subplots(figsize=(8,4))
ax1.plot(d_avGEC[0,:], d_avGEC[1,:])
```

```python
d_S = np.zeros(2*31).reshape(2, 31)
k = 0
for i in range(31):
    d_S[0,i] = i + 1
    d_S[1,i] = sqrt((1./12.) * sum(data.GEC[k:k+12]-d_avGEC[1,i])**2)
    k = k + 12
d_S
```

```python
fig1,ax1=plt.subplots(figsize=(8,4))
ax1.plot(d_S[0,:], d_S[1,:])
```

```python
d_Sss = np.zeros(2*31).reshape(2, 31)
av_monthGEC = mean(data.GEC)
k = 0
for i in range(31):
    d_Sss[0,i] = i + 1
    d_Sss[1,i] = sqrt((1./12.) * sum(data.GEC[k:k+12]-av_monthGEC)**2)
    k = k + 12
d_Sss
```

```python
fig1,ax1=plt.subplots(figsize=(8,4))
ax1.plot(d_Sss[0,:], d_Sss[1,:])
```
