```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)
import pandas as pd
import math
```

```python
data_IRK_m = pd.read_csv('IRKEAGD.DAT', '\s+', parse_dates = True)
data_IRK_m.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
data_IRK_GSM = pd.read_csv('IRKGSMD.DAT', '\s+', parse_dates = True)
data_IRK_GSM.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
data_IRK_exp = pd.read_csv('IRKOBSD.DAT', '\s+', parse_dates = True)
data_IRK_exp.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
```

```python
data_KLD_m = pd.read_csv('KALEAGD.DAT', '\s+', parse_dates = True)
data_KLD_m.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
data_KLD_GSM = pd.read_csv('KALGSMD.DAT', '\s+', parse_dates = True)
data_KLD_GSM.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
data_KLD_exp = pd.read_csv('KALOBSD.DAT', '\s+', parse_dates = True)
data_KLD_exp.columns = ['DaJ', 'UT', 'DaJc', 'foF2', 'dfoF2', 'dfoF2r', 'foF2m']
```

```python
data_IRK_GSM.dfoF2
```

```python
S_IRK_m = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_IRK_m[0,i] = i + 15
    S_IRK_m[1,i] = math.sqrt((1./24.) * sum(data_IRK_m.dfoF2[k:k+24])**2)
    k = k + 24
S_IRK_m

S_IRK_GSM = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_IRK_GSM[0,i] = i + 15
    S_IRK_GSM[1,i] = math.sqrt((1./24.) * sum(data_IRK_GSM.dfoF2[k:k+24])**2)
    k = k + 24
S_IRK_GSM

S_IRK_exp = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_IRK_exp[0,i] = i + 15
    S_IRK_exp[1,i] = math.sqrt((1./24.) * sum(data_IRK_exp.dfoF2[k:k+24])**2)
    k = k + 24
S_IRK_exp
```

```python
S_KLD_m = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_KLD_m[0,i] = i + 15
    S_KLD_m[1,i] = math.sqrt((1./24.) * sum(data_KLD_m.dfoF2[k:k+24])**2)
    k = k + 24
S_KLD_m

S_KLD_GSM = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_KLD_GSM[0,i] = i + 15
    S_KLD_GSM[1,i] = math.sqrt((1./24.) * sum(data_KLD_GSM.dfoF2[k:k+24])**2)
    k = k + 24
S_KLD_GSM

S_KLD_exp = np.zeros(2*17).reshape(2, 17)
k = 0
for i in range(17):
    S_KLD_exp[0,i] = i + 15
    S_KLD_exp[1,i] = math.sqrt((1./24.) * sum(data_KLD_exp.dfoF2[k:k+24])**2)
    k = k + 24
S_KLD_exp
```

```python

fig, ax = plt.subplots(figsize=(12,6))
ax.set_xlim(15,31)
ax.tick_params(labelsize=15)
ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', length=5)
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.set_title('a            Standart deviation for expereimental and model datas from Irkutsk               ', size = 18)
ax.set_ylabel( '\u0394foF2, MHz', size = 15)
ax.set_xlabel('day', size = 15)
ax.axvline(x = 18.5, color = 'k', ls = '--')
p1 = plt.plot(S_IRK_exp[0,:], S_IRK_exp[1,:], 'k--', linewidth = 2.5, label = ' expereimental data')
p2 = plt.plot(S_IRK_m[0,:], S_IRK_m[1,:], 'k', linewidth = 2.5, label = ' EAGLE data')
p3 = plt.plot(S_IRK_GSM[0,:], S_IRK_GSM[1,:], 'k',linewidth = 5, label = ' GSM data')
plt.legend(fontsize = 'x-large')

plt.savefig('IRKEAGnOBSD.png')
```
